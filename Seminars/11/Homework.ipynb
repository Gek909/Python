```python
from sympy import *
from sympy.plotting import plot
from scipy import *
init_printing()
```

```python
x = Symbol('x')
```

```python
f = sin(x)**2 - cos(x)**2
```

# График

```python
plot(f)
```

# Корни

```python
pprint(solveset(f, x, domain=S.Reals), use_unicode=True)
```

# Интервалы, на которых f(x) > 0

```python
pprint(solveset(f > 0, x, domain=S.Reals), use_unicode=True)
```

# Интервалы, на которых f(x) < 0

```python
pprint(solveset(f < 0, x, domain=S.Reals), use_unicode=True)
```

# Производная и интервалы роста и падения

```python
derivative_f = f.diff(x)
derivative_f
```

```python
# 5. Вычислить вершину (найти производню и приравнять к нулю)
pprint(solveset(derivative_f, x, domain=S.Reals), use_unicode=True)
```

```python
# 2. Найти интервалы, на которых функция возрастает (решить неравенства где производная больше или меньше нуля)
pprint(solveset(derivative_f > 0, x, domain=S.Reals), use_unicode=True)
```

```python
# 3. Найти интервалы, на которых функция убывает
pprint(solveset(derivative_f < 0, x, domain=S.Reals), use_unicode=True)
```
